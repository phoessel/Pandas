```python
import pandas as pd
import os
import numpy as np
```

```python
schools_path = os.path.join("C:/", "Resources", "schools_complete.csv")
schools_df = pd.read_csv(schools_path)
students_path = os.path.join("C:/", "Resources", "students_complete.csv")
students_df = pd.read_csv(students_path)
```

```python
complete_df = pd.merge(schools_df,students_df,left_on="name", right_on="school")
complete_df
```

```python
complete_dfbytype = complete_df.loc[complete_df["type"] == "District"]
```

```python
school = complete_dfbytype["school"].nunique()
```

```python
students = complete_dfbytype["name_y"].count()
```

```python
budgetbyschool = complete_dfbytype.drop_duplicates(["name_x"], keep ='first')
budget = budgetbyschool["budget"].sum()
```

```python
av_math_score = complete_dfbytype["math_score"].mean()
```

```python
av_read_score = complete_dfbytype["reading_score"].mean()
```

```python
studentmathpass = complete_dfbytype.loc[(complete_dfbytype["math_score"] >= 60)]
```

```python
passing_number_math = studentmathpass["math_score"].count()
```

```python
percent_passing_math = passing_number_math/students * 100
pass_num_math = round(percent_passing_math, 2)
```

```python
studentreadingpass = complete_dfbytype.loc[(complete_dfbytype["reading_score"] >= 60)]
```

```python
passing_number_reading = studentreadingpass["reading_score"].count()
```

```python
percent_passing_reading = (passing_number_reading/students * 100)
pass_num_reading = round(percent_passing_reading, 2)
```

```python
percent_overall_pass = ((pass_num_math + pass_num_reading) / 2)
overall_pass = round(percent_overall_pass,2)
```

```python
district_df = pd.DataFrame({"Total Schools": [school],"Total Students": [students], "Total Budget": [budget], "Average Math Score": [av_math_score], "Average Reading Score": [av_read_score], "% Passing Math":[pass_num_math], "% Passing Reading":[pass_num_reading], "% Overall Passing Rate": [overall_pass]}) 
```

```python
#reorder the dataframe
df = district_df[["Total Schools","Total Students","Total Budget","Average Math Score","Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing Rate"]]
```

```python
#DISTRICT SUMMARY OUTPUT
df
```

```python
totalstudentsbyschool = complete_df.groupby(["school"])["name_y"].count()
```

```python
countofstudentspassingmath = complete_df.groupby(["school"])["math_score"].agg(lambda val: (val >= 60).sum())
countofstudentspassingreading = complete_df.groupby(["school"])["reading_score"].agg(lambda val: (val >= 60).sum())
```

```python
percentpassreadxschool = countofstudentspassingreading / totalstudentsbyschool * 100
percentpassmathxschool = countofstudentspassingmath / totalstudentsbyschool * 100
percentmath = round(percentpassmathxschool, 2)
percentread = round(percentpassreadxschool, 2)
```

```python
complete_dfbyschool = complete_df.groupby(["school"])
```

```python
total_budget = complete_dfbyschool["budget"].agg(lambda series: series.mode())
```

```python
per_student_budget = total_budget / totalstudentsbyschool
```

```python
agg_math_score = complete_dfbyschool["math_score"].sum()
```

```python
avg_math_score = agg_math_score / totalstudentsbyschool
```

```python
agg_reading_score = complete_dfbyschool["reading_score"].sum()
```

```python
avg_read_score = agg_reading_score / totalstudentsbyschool
```

```python
school_df = complete_dfbyschool.mean()[['size', 'budget', 'reading_score', 'math_score']]
school_df['Per Student Budget'] = school_df.apply(lambda s: s[1] / s[0] , axis=1)
school_df['School Type'] = complete_dfbyschool['type'].agg(lambda s: s.mode())
school_df['cpr'] = complete_df.groupby(["school"])["reading_score"].agg(lambda val: (val >= 60).sum())
school_df['% Passing Reading'] = school_df.apply(lambda s: (s[6] / s[0])*100 , axis=1)
school_df['cpm'] = complete_df.groupby(["school"])["math_score"].agg(lambda val: (val >= 60).sum())
school_df['% Passing Math'] = school_df.apply(lambda s: (s[8] / s[0])*100 , axis=1)
school_df['Overall Passing Rate'] = school_df.apply(lambda s: (s[7] + s[9])/2 , axis=1)
```

```python
school_df.columns = ['Total Students', 'Total School Budget', 'Average Reading Score', 'Average Math Score', 'Per Student Budget', 'School Type', 'cpr', '% Passing Reading', 'cpm', '% Passing Math', 'Overall Passing Rate']
```

```python
#SCHOOL SUMMARY OUTPUT
final_school_df = school_df[["School Type","Total Students","Total School Budget", "Per Student Budget", "Average Math Score","Average Reading Score", "% Passing Math", "% Passing Reading", "Overall Passing Rate"]]
final_school_df.head()
```

```python
#TOP 5 PERFORMING SCHOOLS OUTPUT
topfive_df = final_school_df.sort_values("Overall Passing Rate", ascending = False)
topfive_df.head()
```
